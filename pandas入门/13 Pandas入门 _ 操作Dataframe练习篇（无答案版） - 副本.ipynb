```{"001": {"姓名": "小陈", "考试1": 85, "考试2": 95, "考试3": 92},
 "002": {"姓名": "小李", "考试1": 91, "考试2": 92, "考试3": 94},

 "006": {"姓名": "小周", "考试1": 81, "考试2": 89, "考试3": 92}
 }```
 
 #### 要求索引为学号（001、002等），列名为姓名、考试1、考试2、考试3，并输出创建好的DataFrame。

````python
dic = {"001": {"姓名": "小陈", "考试1": 85, "考试2": 95, "考试3": 92},
 "002": {"姓名": "小李", "考试1": 91, "考试2": 92, "考试3": 94},
 "003": {"姓名": "小王", "考试1": 86, "考试2": 81, "考试3": 89},
 "004": {"姓名": "小张", "考试1": 79, "考试2": 89, "考试3": 95},
 "005": {"姓名": "小赵", "考试1": 96, "考试2": 91, "考试3": 91},
 "006": {"姓名": "小周", "考试1": 81, "考试2": 89, "考试3": 92}}
students = pd.DataFrame(dic)
students
````

````python
students = students.T
students
# students.insert(loc=1,column='年龄',value = [19,19,19,19,19,19],allow_duplicates=True)
# students


````

````python
students['考试4'] = [72, 69, 79, 83, 82, 76]
students

````

#### 给`students`添加一行，索引为007，数据为：`"小杨", 79, 82, 81, 69`。

````python
students.loc['007'] = ["小杨", 79, 82, 81, 69]
students
````

#### 输出`students`删除索引为006、007的结果，但不要改变`students`本身。

````python
students.drop(index=['006','007'],inplace=False)
````

#### 输出`students`删除列名为“考试2”、“考试4”的结果，但不要改变`students`本身。

````python
students.drop(columns=['考试2','考试4'],inplace=False)
````

````python
students
````

#### 创建一个Series`bonus`，包含的值和值对应的索引分别为：`{"考试1": 2, "考试2": 3, "考试3": 2, "考试4": 5}`。输出这个Series与`students`里所有分数列相加后的结果，但不用更新`students`。

````python
bonus_dic = {"考试1": 2, "考试2": 3, "考试3": 2, "考试4": 5}
ser = pd.Series(bonus_dic)
exam = ['考试1','考试2','考试3']
result = students[exam] + ser[exam].values
result
````

#### 更新`students`的`考试4`，让每项数据值增加10，并输出更新后的结果。

````python
students['考试4'] = students['考试4'] + 10
students
````

#### 创建一个叫`student_average`的DataFrame，索引为学号，包含姓名列（列名为“姓名”），以及计算出三次考试平均值后的结果列（列名为“平均分”）。

````python
average_score = students[exam].mean(axis=1)
average_score

student_average = pd.DataFrame({
    '姓名' : students['姓名'],
    '平均分' : average_score}
)
student_average
````
