```python
import pandas as pd
```

```python
f1 = open('./运动员信息表.csv')
df = pd.read_csv(f1)
df
```

```python
data_groupby_project = df.groupby('项目')
data_groupby_project.groups
```

# 统计篮球运动员的平均年龄、身高、体重

```python

basketball_athletes = data_groupby_project.get_group('篮球')
basketball_athletes
```

```python
basketball_athletes_avg = basketball_athletes.agg({
    '年龄（岁）':'mean',
    '身高(cm)':'mean',
    '体重(kg)':'mean'
}).round(1)
basketball_athletes_avg.index = ['平均年龄（岁）','平均身高(cm)','平均体重(kg)']
basketball_athletes = basketball_athletes.assign(**basketball_athletes_avg.to_dict())
basketball_athletes
```

# 统计男运动员的年龄、身高、体重的极差值

```python
def range_data(df):
    return df.max()-df.min()
```

```python
basketball_athletes_groupby_sex = basketball_athletes.groupby('性别')
basketball_man_athletes = basketball_athletes_groupby_sex.get_group('男')
basketball_man_athletes
```

```python
basketball_man_athletes_range = basketball_man_athletes.agg({
    '年龄（岁）':range_data,
    '身高(cm)':range_data,
    '体重(kg)':range_data
})
basketball_man_athletes_range.index = ['平均年龄（岁）','平均身高(cm)','平均体重(kg)']
basketball_man_athletes = basketball_man_athletes.assign(**basketball_man_athletes_range.to_dict())
basketball_man_athletes
```

# 统计篮球运动员的体质指数

```python
def calculate_bmi(df):
    weight = df['体重(kg)']
    height = df['身高(cm)']
    bmi = weight/(height/100)**2
    return bmi
```

```python
basketball_athletes['体质指数'] = (basketball_athletes.apply(calculate_bmi,axis=1)).round(1)
basketball_athletes
```
