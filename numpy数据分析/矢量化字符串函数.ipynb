```python
import pandas as pd
import numpy as np

data = {
    '小红' : '12121212@qq.com',
    '小兰' : '23232323@qq.com',
    '小白' : '12334545www.com',
    '小黄' : np.nan
}
ps = pd.Series(data)
ps
```

```python
print(ps.isnull())
```

```python
ps.map(lambda x:str(x).split('@'))
# ps.map(lambda x:type(x))
```

```python
# 是否包含，返回布尔类型，如果原数据为空则返回nan
ps.str.contains('qq')
```

##### 现查找包含qq的人

```python
# 先处理缺失值
ps2 = ps[ps.notnull()]
ps2.loc[ps2.str.contains('qq')]
```

##### 替换操作

```python
ps.str.replace('qq', 'gamil')
# 可以操作字符串
```

##### 索引

```python
ps.str.get(0)
```

##### 截取

```python
ps.str.slice(0,3)
```

##### 正则表达式

```python
ps.str.extract(r"(\d+)")
```
