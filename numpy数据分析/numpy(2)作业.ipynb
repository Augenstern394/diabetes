### 3、查看下列数组的各项属性（维度，形状，元素个数，类型等）

**<font color="red">关于数组形状,可从外向里看（最外层中包含多少元素项，中间层中每个单元包含多少元素项，最内层每个单元包含多少元素项）</font>**


```python
arr=np.array([[[1,2,3],[4,5,6]],
              [[5,7,8],[7,8,9]],
              [[9,10,11],[12,13,14]]])
arr
```

```python
print('数组arr的维度为:',arr.ndim)
print('数组arr的形状为:',arr.shape)
print('数组arr的元素个数为:',arr.size)
print('数组arr的类型为:',arr.dtype)
```

### 8、尝试利用表达式和函数计算m和n两个数组的加减乘除

```python
m=np.array([[10,20,30],[40,50,60]])
m
```

```python
n=np.array([[1,2,3],[4,5,6]])
n
```

```python
print('m + n = ',np.add(m,n))
print('m - n = ',np.subtract(m,n))
print('m x n = ',np.multiply(m,n))
print('m / n = ',np.divide(m,n))
```
