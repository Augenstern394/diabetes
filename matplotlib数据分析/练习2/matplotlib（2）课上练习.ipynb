```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stu_grade = pd.read_excel('学生成绩表.xlsx')
stu_grade
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

id = stu_grade['学号']
chinese_grade = stu_grade['语文']
fig,ax = plt.subplots(3,2,figsize=(12,11))
ax[0,0].plot(id,chinese_grade,color='r')
ax[0,0].set_title('折线图')
ax[0,0].set_xlabel('学号')
ax[0,0].set_ylabel('语文成绩')

ax[0,1].bar(id,stu_grade['英语'])
ax[0,1].set_title('柱状图')
ax[0,1].set_xlabel('学号')
ax[0,1].set_ylabel('英语成绩')

ax[1,0].scatter(id,stu_grade['数学'])
ax[1,0].set_title('散点图')
ax[1,0].set_xlabel('学号')
ax[1,0].set_ylabel('数学成绩')

ax[1,1].hist(stu_grade['综合'],alpha=0.7)
ax[1,1].set_title('直方图')

ax[2,0].pie(stu_grade['综合'],
            labels=id,
            autopct='%1.1f%%')
ax[2,0].set_title('饼图')


ax[2,1].boxplot(stu_grade['英语'])
ax[2,1].set_title('箱型图')

```

## 练习2

利用如下代码绘制词云图

```python
from PIL import Image
import numpy as np
from wordcloud import WordCloud
import jieba

text = 'Python is great for data analysis and wordcloud is fun'
mask_image = np.array(Image.open('heartshape.png'))

wordcloud = WordCloud(
    font_path='simhei.ttf',
    mask=mask_image,
    background_color='white',
    contour_width=1,
    contour_color='steelblue'
).generate(text)

plt.figure(figsize=(8,6))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```
