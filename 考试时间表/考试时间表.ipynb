```python
import pandas as pd

exam_time_sheet = pd.read_excel('2024--2025学年下学期期末考试日程表.xlsx',header=1)
exam_time_sheet
```

```python
exam_time_sheet.columns
```

```python
filtered_data = exam_time_sheet[
    (exam_time_sheet['年级'] == 2023) & 
    (exam_time_sheet['专业'] == '智能医学工程') &
    (exam_time_sheet['类型'] == '正考')
]
filtered_data
```

```python
with pd.ExcelWriter('2024--2025学年下学期期末考试日程表.xlsx',engine='openpyxl') as writer:
    filtered_data.to_excel(writer,sheet_name='right',index=False)
```
