### pyplot--简单易用，对象：面向当前图

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot([1,2,3],[4,5,6])
plt.show()
```

```python
plt.subplots()
```

### axes--面向对象（面向画布和区域），丰富的接口配置

```python
plt.subplots()
```

```python
fig,ax = plt.subplots()
ax.plot([1,2,3],[4,5,6])
plt.show()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(['北京','上海','广东'],[1,3,5])
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(['北京','上海','广东'],[1,3,5])
plt.show()
```
