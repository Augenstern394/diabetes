```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dic = {
    '小蓝' : 127,
    '小绿' : 127,
    '小黄' : 176,
    '小紫' : 200,
    '小橙' : 154,
    '小粉' : 78,
    '小灰' : 159,
}
dic
```

```python
data = pd.DataFrame({
    '姓名' : list(dic.keys()),
    '成绩总和' : list(dic.values())
})
for i in data['成绩总和']:
    print(i)
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
fig,ax = plt.subplots()
bars = ax.bar(data['姓名'],data['成绩总和'],width=0.4,edgecolor='red',alpha=0.7)
yticks = list(range(0,350,50))
ax.set_yticks(yticks)
ax.set_ylabel('分数',fontsize=12)
ax.set_xlabel('姓名',fontsize=12)
ax.set_title('学生成绩直方图',fontsize=16)

print(dic)
for bar in bars:
    height = bar.get_height()
    print(height)
    print(bar.get_x())
    print(bar.get_width())
    print('=====')
    ax.text(bar.get_x() + bar.get_width() / 2, height, 
            f'{int(height)}', 
            ha='center', va='bottom')

plt.show()
```
