```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
x = ['周一','周二','周三','周四','周五','周六','周日']
y1 = np.random.randint(0,10,size=7)
y2 = np.random.randint(0,5,size=7)
fig,ax = plt.subplots()
ax.bar(x,y1,edgecolor='b',label='系列一')
ax.bar(x,y2,edgecolor='b',label='系列二')
ax.legend(loc=1,fontsize=10)
ax.set_ylabel('y')
ax.set_xlabel('x')
ax.set_title('堆叠柱状图')

for x1,y1,y2 in zip(x,y1,y2):
    print(x1,y1,y2)
    p1 = y1/(y1+y2)
    p2 = y2/(y1+y2)
    plt.text(x1,y1*0.4,'{:.0%}'.format(p1),ha='center',fontsize=16)
    plt.text(x1,y2*0.4,'{:.0%}'.format(p2),ha='center',fontsize=16)
plt.show()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
x = ['周一','周二','周三','周四','周五','周六','周日']
y1 = np.random.randint(0,10,size=7)
y2 = np.random.randint(0,5,size=7)
fig,ax = plt.subplots()
ax.bar(x,y1,edgecolor='b',label='系列一')
ax.bar(x,y2,edgecolor='b',label='系列二',bottom=y1)    # bottom决定谁在下方
ax.legend(loc=1,fontsize=10)
ax.set_ylabel('y')
ax.set_xlabel('x')
ax.set_title('堆叠柱状图')

for x1,y1,y2 in zip(x,y1,y2):
    print(x1,y1,y2)
    p1 = y1/(y1+y2)
    p2 = y2/(y1+y2)
    plt.text(x1,y1*0.4,'{:.0%}'.format(p1),ha='center',fontsize=16)
    plt.text(x1,y1+y2*0.4,'{:.0%}'.format(p2),ha='center',fontsize=16)
plt.show()
```
