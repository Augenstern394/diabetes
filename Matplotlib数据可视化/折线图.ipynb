```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

```python
fix = plt.figure()
ax = fix.add_subplot()
plt.rcParams['font.sans-serif'] = ['SimHei']
x = ['上海', '北京', '昆明']
y = np.array([1,3,5])
ax.plot(x,y,color='r')
plt.show()
```

```python
fix,ax = plt.subplots()
x = ['上海', '北京', '昆明']
y = np.array([1,3,5])
ax.plot(x,y,color='r',linestyle=':') # 默认为实线，--为虚线，-.为长短点线，:为点线  
ax.plot(x,y,'-.k')
```

```python
fig,ax = plt.subplots()
ax.plot([1,2,4],[4,5,6],marker='D')
plt.show()
```

```python

fig,ax = plt.subplots()
ax.plot([1,2,3,4,5],[1,4,9,16,25],alpha=0.5) # alpha设置线条透明度
plt.show()
```

```python

fig,ax = plt.subplots()
x = np.arange(0,10,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
ax.plot(x,y1,label='sin',marker='D',alpha=0.2,linewidth=3,markersize=10)
# linewidth设置线条宽度，markersize设置marker的大小
ax.plot(x,y2,label='cos',alpha=1)
ax.legend()
plt.show()
```
