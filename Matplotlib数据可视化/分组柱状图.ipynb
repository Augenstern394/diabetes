```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
x = list(range(7))
y1 = [7,6,5,4,3,2,1]
y2 = [6,2,3,3,1,1,0.5]
fig,ax = plt.subplots()
ax.bar(x,y1,width=0.4,label='系列1',edgecolor='black',alpha=0.7)
ax.bar([i+0.4 for i in x],y2,width=0.4,label='系列2',edgecolor='black')
ax.set_xlabel('x轴')
plt.xticks(x,['周一','周二','周三','周四','周五','周六','周日'])
ax.set_ylabel('y轴')
ax.set_title('分组柱状图一',fontsize=17)
ax.legend()
plt.show()
```

```python
# 第二种方式
x_arr = np.array(list(range(7)))
y1_arr = [7,6,5,4,3,2,1]
y2_arr = [6,2,3,3,1,1,0.5]
plt.figure(figsize=(9,6))
plt.bar(x_arr,y1_arr,width=0.4,label='系列一',edgecolor='red',alpha=0.7)
plt.bar(x_arr+0.4,y2_arr,width=0.4,label='系列二',edgecolor='blue')
plt.legend(loc=1,fontsize=13)           # loc表示label的位置
plt.title('分组柱状图二',fontsize=16)
plt.xlabel('x轴',fontsize=16)
plt.ylabel('y轴',fontsize=16)

plt.show()
```
