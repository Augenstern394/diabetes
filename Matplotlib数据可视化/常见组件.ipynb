```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 1.设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
# SimHei 微软雅黑
# FangSong 宋体
fig,ax = plt.subplots()
ax.set_title('测试',fontdict={'size':16},loc='center')    # 标题位置loc

ax.plot(['上海','武汉','北京','重庆','昆明','济南'],[10,20,30,40,50,50])
# 设置轴标签
ax.set_xlabel('x轴')
ax.set_ylabel('y轴')
# 设置刻度
ax.set_xticks([0,1,2,3,4,5])
ax.set_yticks([0,10,20,30,40,50])
plt.show()
```

```python
fig,ax = plt.subplots()
# 第一种设置(推荐)：设置图列 注意 label和legend是必须一起使用的，如果缺一则造成图表分不清
ax.plot(['上海','武汉','北京'],[10,20,30],label='2023')
ax.plot(['上海','武汉','北京'],[15,25,35],label='2022')
ax.legend()
plt.show()
```

```python
# 第二种方式
fig,ax = plt.subplots()

ax.plot(['上海','武汉','北京'],[10,20,30])
ax.plot(['上海','武汉','北京'],[15,25,35])
ax.legend([2023,2022])
plt.show()
```

```python
# 设置网格图
fix,ax = plt.subplots()
ax.plot([1,2,3],[4,5,6],color='r')
# 通过color设置线条颜色
ax.grid()
# 如果什么都不写，xy轴都会显示网格线，若需要单一网格则需要设置axis='轴向'
plt.show()
```
