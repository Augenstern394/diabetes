```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
value = list(range(10))
ks = list('qweryfijks')
fig = plt.figure(figsize=(8,4))    # figsize设置宽和高
# 垂直柱状图
ax1 = fig.add_subplot(121) # 1行2列 在第一个
ax1.set_title('图一_垂直图')
ax1.bar(x=ks,
        height=value,
        alpha=0.5,
        width=0.5,
        color = 'y',
        edgecolor='r',
        linewidth=3)
'''
alpha = 透明度
width = 线条宽度
color = 颜色,可以设置多个
edgecolor = 边框颜色
linewidth = 边框宽度
'''


ax2 = fig.add_subplot(122)
ax2.set_title('图二_水平图')
ax2.barh(y=ks,
         width=value,
         alpha=0.5,
         height=0.5,
         color=['green','red','pink'])

plt.show()
```
