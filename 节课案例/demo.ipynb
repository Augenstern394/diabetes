```python
powers = df.groupby('学历')['基本工资'].agg('mean').round(2)
powers
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.bar(powers.index,powers.values,
         color=['green','yellow'],
         alpha=0.5)
plt.xlabel('学历')
plt.ylabel('基本工资')
plt.title('基本工资')
```
