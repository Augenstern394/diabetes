```python
import pandas as pd
```

```python
df_score = pd.read_excel('数据.xlsx',sheet_name='学生成绩表')
df_score
```

```python
total_score = df_score[['数学','英语','计算机']].sum(axis=1)
df_score['总分'] = total_score
result = total_score.map(lambda x: "及格" if x >= 240 else "不及格")
df_score['及格情况'] = result
df_score
```

```python
def query_student_message(id):
    message = df_score.query(f'学号 == {id}')
    return message
```

```python
print(query_student_message('1'))
```
